```python
# MIAS dataset loading

# WANE 

def process_labels(labels, image_size=1024., position_type='float'):
    labels['XMin'] = (labels['X'] - labels['Radius']) / image_size
    labels['XMax'] = (labels['X'] + labels['Radius']) / image_size
    
    labels['YMin'] = (image_size - labels['Y'] - labels['Radius']) / image_size
    labels['YMax'] = (image_size - labels['Y'] + labels['Radius']) / image_size
    labels['Subset'] = 'train'
    return labels

# TODO: Train test split
# make sure annotation is correct 
```

```python
image_dir = 'mias'
img_id = 'mdb001'




def plot_bbox_mias(img_id):
    img_url = get_img_path(img_id)
    img = io.imread(img_url)
    height, width = img.shape
    print(labels[labels["MIAS name"] == img_id])
    bboxs = labels[labels['MIAS name']==img_id]
    for index, row in bboxs.iterrows():
        if row['Abnormality'] == 'NORM':
            print('No abnormality on the image')
            continue
        xmin = row['XMin']
        xmax = row['XMax']
        ymin = row['YMin']
        ymax = row['YMax']
        xmin = int(xmin*width)
        xmax = int(xmax*width)
        ymin = int(ymin*height)
        ymax = int(ymax*height)
        class_name = row['Severity']
        print(f"Coordinates: {xmin,ymin}, {xmax,ymax}")
        cv2.rectangle(img, (xmin,ymin), (xmax,ymax), (255,0,0), 5)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img, class_name, (xmin,ymin-10), font, 3, (0,255,0), 5)
    
    plt.figure(figsize=(15,10))
    plt.title('Image with Bounding Box')
    plt.imshow(img)
    plt.axis("off")
    plt.show()

    
# visualize 10 images WITH tumor
img_ids_for_vis = labels[labels['Severity'].notnull()]["MIAS name"]
for img_id in random.sample(list(img_ids_for_vis), 10):
    plot_bbox_mias(img_id)
    
# labels["MIAS name"].value_counts().tail(50).index.values
# labels[labels['Severity'].notnull()]["MIAS name"]
```

```python
# select train and test subset

# WANE 

def annotate_train_and_test(labels):
    labels = pd.DataFrame(labels)

    # Get unique values from the 'MIAS name' column
    unique_mias_names = labels['MIAS name'].unique()

    # Split the unique values into 80% train and 20% test
    train_names, test_names = train_test_split(unique_mias_names, test_size=0.2, random_state=42)

    # Filter the original DataFrame to create the train and test sets
    labels.loc[labels['MIAS name'].isin(train_names),'Subset'] = 'train'
    labels.loc[labels['MIAS name'].isin(test_names), 'Subset'] = 'test'
    return labels

labels = annotate_train_and_test(labels)
labels[labels['Subset'] == 'test']['Severity'].value_counts()
```

### Write annotation.txt and test_annotation.txt

```python
# WANE 


train_path = 'train'
test_path = 'test'
import math

def write_annotation_txt(directory, annotation_filename, labels, width=1024, heigth=1024):
    with open(os.path.join(directory, annotation_filename), "w+") as f:
        for _, row in labels[labels['Severity'].notnull()].iterrows():
            if math.isnan(row['XMin']):
                continue
            x1 = int(row['XMin'] * width)
            x2 = int(row['XMax'] * width)
            y1 = int(row['YMin'] * heigth)
            y2 = int(row['YMax'] * heigth)
            f.write(row['MIAS name'] + ',' + str(x1) + ',' + str(y1) + ',' + str(x2) + ',' + str(y2) + ',' + row['Severity'] + '\n')            

write_annotation_txt(train_path, 'annotation.txt', labels[labels['subset'] == 'train'])
write_annotation_txt(test_path, 'test_annotation.txt', labels[labels['subset'] == 'test'])
```
